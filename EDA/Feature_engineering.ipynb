# Feature Engineering 
- Age Catogory
- BMI Catogory - Under weight, Normal weight, overweight, Obesity
- Pollution Risk - Loaction and Air pollution Level
- Smoking status
- Interaction Features (somking status and air pollution relation)
- Location Encoding


```python
# import the libary
import pandas as pd 
import numpy as np

```

```python
# load the data 
path = r'C:\Users\kamal\OneDrive\Desktop\COPD_PREDICATION\COPD_PREDICATION\Data\synthetic_COPD_data.csv'
df = pd.read_csv(path)

df.head()
```

```python
# Binning and  discrezation
# Age Catagories


df['Age_Category'] = pd.cut(df['Age'],bins=[29,39,49,59,69,79], labels=['30 - 39','40 - 49','50 - 59','60 - 69','70  - 79']) 

df['BMI_Categories'] = pd.cut(df['BMI'], bins=[0,18.5,24.5,29.5,30], labels=['underweight','Normal weight','Overweight','obessity'])

```

```python
# data Encoding
# Pollution Risk score
df['Pollution_Risk_Score'] = np.where(df['Air_Pollution_Level'] > 150, 1,0)

df['Smoking_Status_encoded'] = df['Smoking_Status'].map({'Current':1 ,'Former':0.5,'Never':0})

# Encode the gender
df['Gender_'] = df['Gender'].map({'Male':1 ,'Female':0})

# Interaction Features - smoking Sstatus and pollution level

df['Smoking_Pollution_interaction'] =df['Smoking_Status_encoded']*df['Air_Pollution_Level']
```
